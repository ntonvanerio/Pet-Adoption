```
main_count = df_train['AdoptionSpeed'].value_counts(normalize=True).sort_index()

def prepare_plot_dict(df, col, main_count):
    main_count = dict(main_count)
    plot_dict = {}
    for i in df[col].unique():
        val_count = dict(df.loc[df[col] == i, 'AdoptionSpeed'].value_counts().sort_index())

        for k, v in main_count.items():
            if k in val_count:
                plot_dict[val_count[k]] = ((val_count[k] / sum(val_count.values())) / main_count[k]) * 100 - 100
            else:
                # If the category is not present, assign -100 to the key 0 in the dictionary.
                plot_dict[0] = -100

    return plot_dict

def make_count_plot(df, x, hue='AdoptionSpeed', title='', main_count=main_count):
    g = sns.countplot(x=x, data=df, hue=hue);
    plt.title(f'AdoptionSpeed {title}');
    ax = g.axes

    plot_dict = prepare_plot_dict(df, x, main_count)

    for p in ax.patches:
        h = p.get_height() if str(p.get_height()) != 'nan' else 0
        # Check if the height 'h' is a key in 'plot_dict'. If not, it means there's no data
        # for that specific category and we assign a default value of -100 to it.
        plot_dict.setdefault(h, -100)
        text = f"{plot_dict[h]:.0f}%" if plot_dict[h] < 0 else f"+{plot_dict[h]:.0f}%"
        ax.annotate(text, (p.get_x() + p.get_width() / 2., h),
             ha='center', va='center', fontsize=11, color='green' if plot_dict[h] > 0 else 'red', rotation=0, xytext=(0, 10),
             textcoords='offset points')
```

```
def plot_four_graphs(col='', main_title='', dataset_title=''):

    plt.figure(figsize=(20, 12));
    plt.subplot(2, 2, 1)
    make_count_plot(df=df_train, x=col, title=f'and {main_title}')

    plt.subplot(2, 2, 2)
    sns.countplot(x='dataset_type', data=all_data, hue=col);
    plt.title(dataset_title);

    plt.subplot(2, 2, 3)
    make_count_plot(df=df_train.loc[df_train['Type'] == 1], x=col, title=f'and {main_title} for dogs')

    plt.subplot(2, 2, 4)
    make_count_plot(df=df_train.loc[df_train['Type'] == 2], x=col, title=f'and {main_title} for cats')

plot_four_graphs(col='Pure_breed', main_title='Pure breed', dataset_title='Number of pure/mixed in train and test data')
```

```
 #variable importance with random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# model prep
df_model = df_train.copy()
X = df_model.drop(columns=['AdoptionSpeed', 'PetID', 'Description', 'Name', 'RescuerID'])
y = df_model['AdoptionSpeed']

for col in X.select_dtypes(include='object').columns:
    X[col] = LabelEncoder().fit_transform(X[col].astype(str))

#  Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

importances = pd.DataFrame({'Feature': X.columns, 'Importance': model.feature_importances_})
importances = importances.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=importances.head(15))
plt.title('Top 15 Variables from Random Forest')
plt.tight_layout()
plt.show()
```

```
import lightgbm as lgb
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from lightgbm import early_stopping, log_evaluation
# Target: AdoptionSpeed
y = df_train['AdoptionSpeed']

# Features
X = df_train.drop(columns=['AdoptionSpeed', 'PetID', 'Name', 'Description', 'RescuerID'])

#
for col in X.select_dtypes(include='object').columns:
    X[col] = LabelEncoder().fit_transform(X[col].astype(str))


# split train/test
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Dataset  LightGBM
lgb_train = lgb.Dataset(X_train, label=y_train)
lgb_val = lgb.Dataset(X_val, label=y_val, reference=lgb_train)

# basic parametres
params = {
    'objective': 'multiclass',
    'num_class': y.nunique(),
    'metric': 'multi_logloss',
    'learning_rate': 0.05,
    'verbose': -1
}

# training with callbacks
lgb_model = lgb.train(
    params,
    lgb_train,
    valid_sets=[lgb_val],
    num_boost_round=100,
    callbacks=[
        early_stopping(stopping_rounds=10),
        log_evaluation(period=10)
    ]
)
lgb_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': lgb_model.feature_importance(importance_type='gain')
}).sort_values(by='Importance', ascending=False)

# graph
plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=lgb_importance.head(15))
plt.title('Top 15 Variables importance by LightGBM')
plt.tight_layout()
plt.show()
```
